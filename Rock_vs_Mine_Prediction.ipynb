Importing the dependency

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

Data Collection and data processing

```python
#loading the datasets to pandas Dataframe
sonar_data = pd.read_csv('/content/sonar data.csv',header= None)
```

```python
sonar_data.head()
```

```python
#number of rows and column
sonar_data.shape
```

```python
sonar_data.describe() #it gives stastical measure of the data
```

```python
sonar_data[60].value_counts()
```

M->mine
R->rock

```python
sonar_data.groupby(60).mean()
```

```python
#seprating data and label
x = sonar_data.drop(columns=60,axis=1)
y = sonar_data[60]
```

```python
print(x)
print(y)
```

Training and test data

```python
X_train, X_test , Y_train, Y_test  = train_test_split(x,y,test_size = 0.1, stratify=y,random_state=1)
```

```python
print(x.shape,X_train.shape,X_test.shape)
df = pd.DataFrame({
    'Name': ['Total','Data', 'Testing',],
    'Sets': [x.shape[0],X_train.shape[0],X_test.shape[0]]
})

# plotting a bar graph
df.plot(x="Name", y="Sets", kind="bar")
```

```python
print(X_train)
print(Y_train)
```

Model training -> Logistic regression function

```python
model = LogisticRegression()
```

```python
#training the Logistic regression model with training data
model.fit(X_train,Y_train)
```

Model Evaluation

```python
#accuracy on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data : ',training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data : ',test_data_accuracy)
```
